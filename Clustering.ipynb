### Checking class to areas

Antes que esto se hace el vic para encontrar la mejor manera de clasificar el clustering para interpretarlo, y para saber cual clustering es el mejor de los dos métodos. 

1 is a zero damage crash

0 is a low or higher damage crash

```
filex = 'gdrive/My Drive/DATLAS/DF_UNIFICADO.csv'
df = pd.read_csv(filex)
df['class'] = labels
df = df.drop(['Unnamed: 0', 'Unnamed: 0.1', 'FOLIO_ID', 'complete_date', 'date_time','gh4', 'gh5', 'gh6', 'gh8', 'AÑO', 'MES', 'DÍA NUMERO', 'DIA', 'ESTADO',
       'CIUDAD_APROXIMADA','moonrise', 'moonset', 'sunrise', 'sunset'],axis=1)
```

```
df.head()
```

```
gh_vals = df['gh7'].value_counts()
print(gh_vals)
```

```
classcount = {}

for index, row in df.iterrows():
  if df['class'][index] == 1:
    try:
      classcount[df['gh7'][index]] += 1
    except:
      classcount[df['gh7'][index]] = 1
  else:
    try:
      classcount[df['gh7'][index]] += 0
    except:
      classcount[df['gh7'][index]] = 0
```

```
gh_proportion = {}
# Proportion of zerodamage/total_crashes
for gh in gh_vals.keys():
  gh_proportion[gh] = classcount[gh]/gh_vals[gh]
```
