```
df['date_time']= pd.to_datetime(df['date_time']) 
```

```
import datetime

df['night'] = ''

for index, row in df.iterrows():
  dtime = pd.to_datetime(row['sunrise'], format='%I:%M %p')
  ntime = pd.to_datetime(row['sunset'], format='%I:%M %p')

  if row['date_time'].hour >= dtime.hour and row['date_time'].hour < ntime.hour:
    df['night'][index] = 0
  else:
    df['night'][index] = 1
```
